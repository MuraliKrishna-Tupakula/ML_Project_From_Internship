## Diamond price prediction

#### Colour, Clarity, Carat (size), Cut, Shape and Fluorescence are the major determinants of the price of diamonds. 
Difference in single grade or level will make the price change from 5% to 30%

#### Diamond dealers compete to be one of 84 “sightholders” at De Beers' “sights.” 
These companies are invited to view and purchase diamonds directly from De Beers ten times a year.
De Beers determines how much and what kind of diamond rough to sell to each company or sightholder and at what price.

```python
X = Di_pred_price_Upd['carat'].values.reshape(-1,1)
y = Di_pred_price_Upd['price'].values.reshape(-1,1)
X
```

```python
#Ploting a graph to see the points
plt.figure(figsize=(16, 8))
plt.scatter(X, y, c='black')
plt.xlabel("carat")
plt.ylabel("price")
plt.show()
```

```python
#Splitting our dataset to Training and Testing dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
X_test
```

```python
X_train
```

### LINEAR REGRESSION  

```python
#Fitting Linear Regression to the training set
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
y_train[:,:5]
y_test[:,:5]
#X_test[:,:5]
```

```python
y_pred=reg.predict(X_test)
y_pred[:5]
```

#### output how much accuracy got on our prepared model 

```python
#Checking the Accuracy
from sklearn.metrics import r2_score,accuracy_score
r2_score(y_test,y_pred)
```

```python
#predicting the Test set result
y_pred = reg.predict(X_test)
plt.figure(figsize=(20,10))
plt.scatter(X, y, c='black')
plt.plot(
    X_test,
    y_pred,
    c='blue',
    linewidth=5
)
plt.xlabel("Money spent based on carat ")
plt.ylabel("Diamond cost")
plt.title("Prediction Points vs Actual Points")
plt.show()
```
